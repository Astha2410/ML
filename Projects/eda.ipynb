```python
import plotly.graph_objs as go

```

```python
data = go.Scatter(
    x= chl.date,
    y= chl.Chlorophyll,
    mode="markers")
```

```python
layout = go.Layout(
    title = "chlorohyll visuallisation", 
    xaxis = dict(title = "dates"),
    yaxis = dict(title = "chorophyll values")

)
```

```python
fiqure= go.Figure(data= data, layout= layout)
```

```python
pyo.plot(fiqure)
```

```python
trace = go.Histogram(
    x = chl.Chlorophyll,
    name = "chlorophyll",
    opacity = .5
)
```

```python
data = [trace]
layout = go.Layout(title = "chlorophyll visualisation",
                   xaxis = dict(title = "chlorophyll values"),
                   yaxis = dict(title = "count"))
```

```python
fig = go.Figure(data, layout = layout)
```

```python
pyo.plot(fig)
```
